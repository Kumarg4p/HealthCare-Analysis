```python
# As per description below five column values zeros considered as missing values
# Glucose, BloodPressure, SkinThickness, Insulin, BMI
# replacing zeros on thos fields with Nan

df_HC['Glucose'].loc[df_HC['Glucose'] == 0] = np.nan
df_HC['BloodPressure'].loc[df_HC['BloodPressure'] == 0] = np.nan
df_HC['SkinThickness'].loc[df_HC['SkinThickness'] == 0] = np.nan
df_HC['Insulin'].loc[df_HC['Insulin'] == 0] = np.nan
df_HC['BMI'].loc[df_HC['BMI'] == 0] = np.nan
```

```python
#value Zeros are converted to null values
df_HC.info()
```

```python
# checking replaced with Nan
df_HC.tail()
```

```python
#Handling missing values 
#Glucose, BloodPressure, SkinThickness, Insulin, BMI - Mean (since all are numeric)
df_HC['Glucose'].fillna(int(df_HC.Glucose.mean()) , inplace=True)
df_HC['BloodPressure'].fillna(int(df_HC.BloodPressure.mean()) , inplace=True)
df_HC['SkinThickness'].fillna(int(df_HC.SkinThickness.mean()) , inplace=True)
df_HC['Insulin'].fillna(int(df_HC.Insulin.mean()) , inplace=True)
df_HC['BMI'].fillna(int(df_HC.BMI.mean()) , inplace=True)
df_HC.info()
```

```python
#Write CSV after handling null values, it can be used in Tableau
df_HC.to_csv('HealthCare.csv')
```

# Week 3

```python
# separate the date is features and label
# Features 2D
# Label 1D

features = df_HC.iloc[:,:-1].values
label = df_HC.iloc[:,-1].values
```

# Logistic Regression

```python
#Logistic regression
# get the best random_state for generalized model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
for i in range(1,769):
    X_train, X_test, y_train, y_test = train_test_split(features,
                                                   label,
                                                   test_size =0.2,
                                                   random_state=i)
    model = LogisticRegression()
    model.fit(X_train,y_train)
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    
    if test_score > train_score:
        print("Testing {} Training {} Random_state {}".format(test_score, train_score, i))
```

```python
# Train the model
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# Check Generalization
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

# KNN 

```python
#KNN
# get the best random_state and K value for generalized model
# Looped random_State for record value count 768
# K value from 1 to 600
# The best score found - Testing 0.8636363636363636 Training 0.745928338762215 Random_state 506 K_value 67
# KNN is best compare to Logistic regression score 0.837662337662
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
for i in range(1,769):
    for j in range(1,101):
        X_train, X_test, y_train, y_test = train_test_split(features,
                                                   label,
                                                   test_size =0.2,
                                                   random_state=i)
        model = KNeighborsClassifier(n_neighbors = j)
        model.fit(X_train,y_train)
        train_score = model.score(X_train, y_train)
        test_score = model.score(X_test, y_test)
    
        if test_score > train_score and test_score >= 0.8636363636363636:
            print("Testing {} Training {} Random_state {} K_value {}".format(test_score, train_score, i, j))
```

# SVM

```python
#SVM
# get the best random_state to generalized model
# Looped random_State for record value count 768

from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
for i in range(1,769):
    X_train, X_test, y_train, y_test = train_test_split(features,
                                                   label,
                                                   test_size =0.2,
                                                   random_state=i)
    model = SVC(gamma=0.4, kernel='rbf')
    model.fit(X_train,y_train)
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    
    if test_score >= train_score:
        print("Testing {} Training {} Random_state {}".format(test_score, train_score, i))
```

# Xgboost

```python
# get the best random_state to generalized model
# Looped random_State for record value count 768
#Best score got on xgb is Testing : 0.8636363636363636 Training 0.7801302931596091 RandomState 506
#The same score got in KNN algorithm
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
for i in range(1,769):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                label,
                                                test_size=0.2,
                                                random_state=i)
    model = XGBClassifier(learning_rate=0.01)
    model.fit(X_train,y_train)
    
    train_score=model.score(X_train,y_train)
    test_score=model.score(X_test,y_test)
    
    if test_score > train_score:
        print("Testing : {} Training {} RandomState {}".format(test_score,train_score,i))
```

# Decision Tree

```python
# get the best random_state to generalized model
# Looped random_State for record value count 768
#Best score got on Decision Tree is Testing : 0.8246753246753247 Training 0.758957654723127 RandomState 395
#This score comparatively less than KNN and Xgb algorithms
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
for i in range(1,769):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                label,
                                                test_size=0.2,
                                                random_state=i)
    model = DecisionTreeClassifier(max_depth=3)
    model.fit(X_train,y_train)
    
    train_score=model.score(X_train,y_train)
    test_score=model.score(X_test,y_test)
    
    if test_score > train_score:
        print("Testing : {} Training {} RandomState {}".format(test_score,train_score,i))
```

# Random Forest

```python
# get the best random_state to generalized model
# Looped random_State for record value count 768
#Best score got on Random Forest is Testing : 0.8246753246753247 Training 0.758957654723127 RandomState 395
#This score comparatively less than KNN and Xgb algorithms
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
for i in range(1,769):
    X_train,X_test,y_train,y_test = train_test_split(features,
                                                label,
                                                test_size=0.2,
                                                random_state=i)
    model = RandomForestClassifier(n_estimators=100, max_depth=67)
    model.fit(X_train,y_train)
    
    train_score=model.score(X_train,y_train)
    test_score=model.score(X_test,y_test)
    
    if test_score > train_score:
        print("Testing : {} Training {} RandomState {}".format(test_score,train_score,i))
```

```python
#Print Confusion Matrix for entire dataset
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(label, model.predict(features))
print(cm)
```

```python
#Since dataset is unbalanced
from sklearn.metrics import classification_report
cr = classification_report(label, model.predict(features))
print(cr)
```

```python
#ROC curve is used for Balanced dataset
#Since this unbalanced dataset we can use PR curve
# I have used xgboost model algo that has more accuracy (KNN also the same accuracy)
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
```

```python
#Get The probabilities score
probs = model.predict_proba(features)
```

```python
#Keep the positive outcomes ---- Keep True Probabilities
#first one is probability of 0 and second is the probability of 1 (positive)
probs = probs[:,1]
```

```python
#Perform prediction of features
y_pred = model.predict(features)
y_pred
```

```python
#Cal PR score
precision, recall, thre = precision_recall_curve(label,probs)
```

```python
#Cal F1 score
f1 = f1_score(label,y_pred)
```

```python
#Calc AUC
# x-axis is recall and y axis is precision
auc = auc(recall,precision)
```

```python
#Plot PR with Skill line
plt.plot([0,1],[0.5,0.5], linestyle='--')
plt.plot(recall,precision, label='PR Curve(Area = %0.2f)' %auc)
plt.legend(loc = "lower right")

print(auc)
```
